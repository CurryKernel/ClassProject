```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk 
from nltk.tokenize import WordPunctTokenizer
```

```python
f = open("Reviews.txt")
raw = f.readlines()
text = raw[0]
text
```

## 分割成词语，去除标点

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
words = tokenizer.tokenize(raw[0])
words
```

## 打标记 

```python
tags = nltk.pos_tag(words)
tags
```

## stemming

```python
# from nltk.stem import SnowballStemmer
# snowball_stemmer = SnowballStemmer('english')
# word = []
# for w in words:
#     word.append(snowball_stemmer.stem(w))
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
WORD = []

for word, tag in tags:
     wntag = tag[0].lower()
     wntag = wntag if wntag in ['a', 'r', 'n', 'v'] else None
     if not wntag:
             lemma = word
     else:
             lemma = wordnet_lemmatizer.lemmatize(word, wntag)
     WORD.append(lemma)
WORD
```

## stop word

```python
from nltk.corpus import stopwords
word_after_stop = []
for w in WORD:
    if not w in stopwords.words('english'):
        word_after_stop.append(w)
word_after_stop
```

## 去除过短的词和spell check（conti.）

```python
clean=""
for w in word_after_stop:
    if len(w)>=3 and w.isalpha():
        clean = clean + w.lower() +" "
clean = clean[:-1]
clean
all=[]
all.append(clean)
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(min_df=1)
vectorizer.fit_transform(all)
```
