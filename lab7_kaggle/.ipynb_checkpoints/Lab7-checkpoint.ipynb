## 3）离散特征 使用one-hot方法处理

```python
dummies_Cabin = pd.get_dummies(data['Cabin'],prefix='Cabin')
dummies_Sex = pd.get_dummies(data['Sex'],prefix='Sex')
dummies_Pclass = pd.get_dummies(data['Pclass'],prefix='Pclass')
dummies_Embarked = pd.get_dummies(data['Embarked'],prefix='Embarked')
dummy_Parch = pd.get_dummies(data.Parch, prefix='Parch')
dummy_SibSp = pd.get_dummies(data.SibSp, prefix='SibSp')
dummy_Age = pd.get_dummies(data.Age, prefix='Age')

data = pd.concat([data,dummies_Cabin,dummies_Sex,dummies_Pclass,dummies_Embarked,dummy_Parch,dummy_SibSp,dummy_Age],axis=1)
data.head(10)
```

# 4. 模型预测

```python
from sklearn.ensemble import GradientBoostingClassifier   
from sklearn.linear_model.logistic import LogisticRegression
from sklearn import tree
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold

data_np = data.values

test_np = test.values

Y = data_np[:,1]

X = data_np[:,2:]
print(X)

Z = test_np[:,1:]
print(Z)
```

## 计算准确率

```python
def get_acc(predict, target):
    totle = len(predict)
    trueCount = 0
    for i in range(totle):
        if predict[i] == target[i]:
            trueCount += 1
    return float(trueCount) / float(totle)
```

```python
accList = []
skf = StratifiedKFold(n_splits=10)  
model = GradientBoostingClassifier(n_estimators=200,learning_rate=0.05)  
#clf = LogisticRegression()
#clf = tree.DecisionTreeClassifier()
clf = SVC()
for train,validation in skf.split(X,Y):
    clf.fit(X[train],Y[train])
    labelPredict = clf.predict(X[validation])
    accList.append(get_acc(labelPredict,Y[validation]))
    
averangACC = sum(accList)/len(accList)
print(averangACC)
```

# 5.测试数据

```python
test_target_value = clf.predict(Z)
```

```python
result_df = pd.DataFrame(columns=['PassengerId', 'Survived'])  
result_df['PassengerId'] = test['PassengerId']  
result_df['Survived'] = test_target_value.astype(int)
result_df.to_csv('MyPredict.csv', index=False, header=['PassengerId', 'Survived'])  
```
